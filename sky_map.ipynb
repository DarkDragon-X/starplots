```python
import math
import random
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df_bsc = pd.read_csv('bsc5.csv')
long=[]
lat = []
for i in range(len(df_bsc)):
    ra_h = df_bsc.at[i, "RAh"]
    ra_m = df_bsc.at[i, "RAm"]
    ra_s = df_bsc.at[i, "RAs"]

    if pd.notnull(ra_h) and pd.notnull(ra_m) and pd.notnull(ra_s):
        RaDeg=math.radians(
            (int(ra_h) / 24) * 360 +
            (int(ra_m) / 60) * 15 +
            (int(ra_s) / 3600) * 15)
        if RaDeg >= 0 and RaDeg <= math.pi:
            long.append(RaDeg)
        elif RaDeg > math.pi and RaDeg <= 2 * math.pi:
            long.append(RaDeg - (2 * math.pi))
    
    dec_sign = df_bsc.at[i, "DE-"]
    dec_deg = df_bsc.at[i, "DEd"]
    dec_min = df_bsc.at[i, "DEm"]
    dec_sec = df_bsc.at[i, "DEs"]
    
    if pd.notnull(dec_sign) and pd.notnull(dec_deg) and pd.notnull(dec_min) and pd.notnull(dec_sec):
        declination = dec_deg + (dec_min / 60) + (dec_sec / 3600)
        if dec_sign == '-':
            declination = -declination
        lat.append(math.radians(declination))
```

```python
plt.figure (figsize = (16, 8.4))
plt.subplot (projection = "aitoff")
plt.title ("Map of the Sky")
plt.plot (long, lat, 'o', markersize = 1, alpha = 0.4, color = 'r')
plt.grid (True)
plt.savefig ('sky_map.pdf')
plt.show()
```

```python
df_bsc = pd.read_csv('bsc5.csv')
long=[]
lat = []
for i in range(len(df_bsc)):
    glon = df_bsc.at[i, "GLON"]
    glat = df_bsc.at[i, "GLAT"]

    if pd.notnull(glon) and pd.notnull(glat):
        glon=math.radians(glon)
        glat=math.radians(glat)
        if glon >= 0 and glon <= math.pi:
            long.append(glon)
        elif glon > math.pi and glon <= 2 * math.pi:
            long.append(glon - (2 * math.pi))
        lat.append(glat)
```

```python
plt.figure (figsize = (16, 8.4))
plt.subplot (projection = "aitoff")
plt.title ("Map of the Sky")
plt.plot (long, lat, 'o', markersize = 1, alpha = 0.4, color = 'r')
plt.grid (True)
plt.savefig ('sky_map.pdf')
plt.show()
```
